# Support Vector Machines - Exercise 1

In this exercise, we'll be using support vector machines (SVMs) to build a spam classifier.  We'll start with SVMs on some simple 2D data sets to see how they work.  Then we'll do some pre-processing work on a set of raw emails and build a classifier on the processed emails using a SVM to determine if they are spam or not.

The first thing we're going to do is look at a simple 2-dimensional data set and see how a linear SVM works on the data set for varying values of C (similar to the regularization term in linear/logistic regression).  Let's load the data.
## Exercise 1
#### 1. Load libraries

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

#### 2. Load data
Load the file *ejer_1_data1.mat*. Find the way for loading this kind of file. **scipy.io.loadmat**

````python
import scipy.io
mat = scipy.io.loadmat('data/ejer_1_data1.mat')
````

````python
mat.keys()
````

````python
mat['X']
````

````python
mat['y']
````

#### 3. Create a DataFrame with the features and target

````python
# Crear DataFrame con las features
df = pd.DataFrame(mat["X"], columns=["x1", "x2"])

# Agregar la columna de la variable objetivo (target)
df["target"] = mat["y"]

# Verificar primeras filas
df
````

#### 4. Plot a scatterplot with the data

````python
# Creamos figura
plt.figure(figsize=(8, 6))

# Clase 0 → color rojo
plt.scatter(df[df["target"] == 0]["x1"], df[df["target"] == 0]["x2"], 
            color="red", label="Clase 0", edgecolor="k")

# Clase 1 → color azul
plt.scatter(df[df["target"] == 1]["x1"], df[df["target"] == 1]["x2"], 
            color="blue", label="Clase 1", edgecolor="k")

# Estética del gráfico
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatterplot de las clases")
plt.legend()
plt.grid(True)
plt.show()
````

Notice that there is one outlier positive example that sits apart from the others.  The classes are still linearly separable but it's a very tight fit.  We're going to train a linear support vector machine to learn the class boundary.

#### 5. LinearSVC
Declare a Linear SVC with the hyperparamenters:

```Python
LinearSVC(C=1, loss='hinge', max_iter=10000)
```

````python
from sklearn.svm import LinearSVC

# Instanciamos el modelo
modelo_svm = LinearSVC(C=1, loss='hinge', max_iter=10000)

# Entrenamos el modelo con las features x1 y x2
X = df[["x1", "x2"]]
y = df["target"].values.ravel()  # Aseguramos que y tenga la forma correcta

modelo_svm.fit(X, y)
````

#### 6. Try the performance (score)
For the first experiment we'll use C=1 and see how it performs.

````python
# Evaluamos el rendimiento del modelo SVM en los datos de entrenamiento
score = modelo_svm.score(X, y)

# Mostramos el resultado
print("Accuracy del modelo (C=1):", score)
````

It appears that it mis-classified the outlier.

#### 7. Increase the value of C until you get a perfect classifier

````python
from sklearn.svm import LinearSVC

# Valores de C para probar
valores_C = [1, 10, 100, 1000,10000]

# Variables de features y target (reusamos lo que ya tenés)
X = df[["x1", "x2"]]
y = df["target"].values.ravel()

for c in valores_C:
    modelo = LinearSVC(C=c, loss='hinge', max_iter=10000)
    modelo.fit(X, y)
    score = modelo.score(X, y)
    print(f"Accuracy con C={c}: {score}")
````

This time we got a perfect classification of the training data, however by increasing the value of C we've created a decision boundary that is no longer a natural fit for the data.  We can visualize this by looking at the confidence level for each class prediction, which is a function of the point's distance from the hyperplane.

#### 8. Plot Decission Function
Get the `decision_function()` output for the first model. Plot a scatterplot with X1, X2 and a range of colors based on `decision_function()`

````python
import matplotlib.pyplot as plt

# 1. Obtener las distancias al hiperplano
decision_values = modelo_svm.decision_function(X)

# 2. Hacer scatterplot con color según esas distancias
plt.figure(figsize=(8,6))
scatter = plt.scatter(X["x1"], X["x2"], c=decision_values, cmap='coolwarm', edgecolor='k')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Decision Function: Confianza del modelo")
plt.colorbar(scatter, label='Distancia al hiperplano')
plt.grid(True)
plt.show()
````

For this data set we'll build a support vector machine classifier using the built-in RBF kernel and examine its accuracy on the training data.  To visualize the decision boundary, this time we'll shade the points based on the predicted probability that the instance has a negative class label.  We'll see from the result that it gets most of them right.

#### 4. Declare a SVC with this hyperparameters
```Python
SVC(C=100, gamma=10, probability=True)
```


````python
from sklearn.svm import SVC

# Features y target
x = df2[["1", "2"]]
y = df2["target"].values.ravel()

# Instanciar el modelo con los parámetros indicados
modelo_rbf = SVC(C=100, gamma=10, probability=True)

# Entrenar
modelo_rbf.fit(X, y)
````

#### 5. Fit the classifier and get the score

````python
# Ajustamos el modelo (en caso de que no lo hayas hecho todavía)
modelo_rbf.fit(x, y)

# Obtenemos el score en el conjunto de entrenamiento
score = modelo_rbf.score(x, y)

# Mostramos el resultado
print("Accuracy del modelo RBF:", score)
````

#### 6. Plot the scatter plot and probability of predicting 0 with a [sequential color](https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html)

````python
import matplotlib.pyplot as plt

# Obtenemos las probabilidades de predicción para clase 0
prob_clase_0 = modelo_rbf.predict_proba(X)[:, 0]  # Probabilidad de que cada punto sea clase 0

# Hacemos el scatterplot
plt.figure(figsize=(8,6))
scatter = plt.scatter(x["1"], x["2"], c=prob_clase_0, cmap="Greens", edgecolor='k')
plt.colorbar(scatter, label="Probabilidad de clase 0")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Predicción de clase 0 (probabilidad) con color secuencial")
plt.grid(True)
plt.show()
````
