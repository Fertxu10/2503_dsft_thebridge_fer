2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
import matplotlib.pyplot as plt
import numpy as np

# CREAMOS EL EJE X
x = np.arange(0, 5, 0.5)
```

```python
# CREAMOS LAS 2 FIGURAS

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
```

```python

# CREAMOS LAS 2 FIGURAS
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# RECTA CON PENDIENTE +3 Y PASA POR (0,0)
y1 = 3 * x
ax1.plot(x, y1, color='blue')
ax1.set_title("Recta con pendiente +3\n(pasa por 0,0)")
ax1.grid(True)

# RECTA CON PENDIENTE 33 Y PASA POR (0,-5)
y2 = -3 * x - 5
ax2.plot(x, y2, color='red')
ax2.set_title("Recta con pendiente –3\n(pasa por 0,–5)")
ax2.grid(True)

# MOSTRARLO
plt.tight_layout()
plt.show()
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5.
```python
x = np.arange(0, 5, 0.5)

fig,axis = plt.subplots(1, 2, figsize=(10, 4))


y1 = 3 * x
axis[0].plot(x, y1, color='blue')
axis[0].set_title("Recta con pendiente +3\n(pasa por 0,0)")
axis[0].set(xlim=(0, 5), ylim=(-15, 15), xlabel="Eje X", ylabel="Eje Y", title="Recta sube")
axis[0].grid(True)


y2 = -3 * x - 5
axis[1].plot(x, y2, color='red')
axis[1].set_title("Recta con pendiente –3\n(pasa por 0,–5)")
axis[1].set(xlim=(0, 5), ylim=(-15, 15), xlabel="Eje X", ylabel="", title="Recta baja")
axis[1].grid(True)


plt.tight_layout()
plt.show()
```

7. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es más sencillo dejarlos dentro de la figura.

```python

# Datos
nombres = ["Ana", "Luis", "Pedro", "Sofía", "Carmen"]
alturas = [160, 180, 175, 190, 170]

# Crear figura con ejes integrados (lo más simple)
plt.figure(figsize=(8, 5))

# Pintar el gráfico de barras en color verde
plt.bar(nombres, alturas, color="green")
plt.ylim(100, 200)

# Añadir etiquetas y título
plt.title("Altura de amigos")
plt.xlabel("Amigos")
plt.ylabel("Altura (cm)")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Datos
amigos = {"Ana" : 160,
          "Luis" : 180,
          "Pedro" : 175,
          "Sofía" : 190,
          "Carmen" : 170}

# Crear figura con ejes integrados (lo más simple)
plt.figure(figsize=(8, 5))

# Pintar el gráfico de barras en color verde
plt.bar(amigos.keys(), amigos.values(), color="green")
plt.ylim(100, 200)

# Añadir etiquetas y título
plt.title("Altura de amigos")
plt.xlabel("Amigos")
plt.ylabel("Altura (cm)")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

11. Prueba a invertir los ejes con ax.invert_axis()
Necesitarás tener los ejes disponibles fuera de la figura

```python
# Datos
nombres = ["Ana", "Luis", "Pedro", "Sofía", "Carmen"]
alturas = [160, 180, 175, 190, 170]

# Crear figura y eje
fig, ax = plt.subplots(figsize=(8, 5))
plt.style.use('seaborn-v0_8')

# Dibujar barras
ax.barh(nombres, alturas, color="green")
plt.xlim(100, 200)

# Añadir etiquetas
ax.set_title("Altura de un grupo de amigos")
ax.set_xlabel("Persona")
ax.set_ylabel("Altura (cm)")

# Invertir el eje X
ax.invert_xaxis()
# Invertir el eje Y
ax.invert_yaxis()

# Mostrar
plt.tight_layout()
plt.show()
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
df = pd.DataFrame({
    "Year": [2015, 2016, 2017, 2018, 2019],
    "Sold_items_A": [1000, 3500, 4000, 5500, 7000],
    "Sold_items_B": [2000, 3100, 5000, 4000, 6000]
})

#plt.style.use("seaborn-white")


plt.plot(df["Year"], df["Sold_items_A"], linestyle="--", color="green", label="Departamento A")
plt.plot(df["Year"], df["Sold_items_B"], linestyle=":", color="red", label="Departamento B")

plt.title("Ventas por Año")
plt.xlabel("Año")
plt.ylabel("Unidades Vendidas")


plt.legend(loc="lower right")


plt.tight_layout()
plt.show()
```
